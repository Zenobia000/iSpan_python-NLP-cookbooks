```python
from torch.utils.tensorboard import SummaryWriter
```

```python
writer = SummaryWriter("logs")

# writer.add_image()
# y = x
for i in range(100):
    writer.add_scalar("y=3x", 3*i, i)

writer.close()
```

```python
# launch tensor bard 
%tensorboard --logdir=logs --port=6007
```
