````python
# Ë¶ñË¶∫Âåñ NLP ÁôºÂ±ïÊôÇÈñìËª∏
import matplotlib.patches as mpatches

fig, ax = plt.subplots(figsize=(16, 10))
ax.set_xlim(1950, 2025)
ax.set_ylim(0, 10)
ax.axis('off')

# ÂõõÂ§ßÈöéÊÆµÊôÇÈñìËª∏
stages = [
    {'name': 'Ë¶èÂâáËàáÁ¨¶ËôüÊôÇ‰ª£', 'start': 1950, 'end': 1990, 'color': '#FFB6C1', 'y': 8},
    {'name': 'Áµ±Ë®àÊ©üÂô®Â≠∏ÁøíÊôÇ‰ª£', 'start': 1990, 'end': 2013, 'color': '#87CEEB', 'y': 6},
    {'name': 'Ê∑±Â∫¶Â≠∏ÁøíÊôÇ‰ª£', 'start': 2013, 'end': 2017, 'color': '#98FB98', 'y': 4},
    {'name': 'Transformer ÊôÇ‰ª£', 'start': 2017, 'end': 2024, 'color': '#FFD700', 'y': 2}
]

for stage in stages:
    rect = mpatches.Rectangle((stage['start'], stage['y']-0.4), stage['end']-stage['start'], 0.8,
                               linewidth=2, edgecolor='black', facecolor=stage['color'], alpha=0.7)
    ax.add_patch(rect)
    ax.text((stage['start']+stage['end'])/2, stage['y'], stage['name'], 
            ha='center', va='center', fontsize=13, fontweight='bold')

# ÈóúÈçµÈáåÁ®ãÁ¢ë
milestones = [
    {'year': 1966, 'event': 'ELIZA', 'y': 8.8},
    {'year': 2003, 'event': 'Word2Vec', 'y': 6.8},
    {'year': 2014, 'event': 'Seq2Seq', 'y': 4.8},
    {'year': 2017, 'event': 'Transformer', 'y': 2.8},
    {'year': 2018, 'event': 'BERT', 'y': 2.5},
    {'year': 2020, 'event': 'GPT-3', 'y': 2.2},
    {'year': 2022, 'event': 'ChatGPT', 'y': 1.9},
    {'year': 2023, 'event': 'GPT-4', 'y': 1.6},
]

for milestone in milestones:
    ax.plot([milestone['year'], milestone['year']], [0.5, milestone['y']], 
            'k--', linewidth=1.5, alpha=0.6)
    ax.scatter(milestone['year'], milestone['y'], s=200, c='red', zorder=5, edgecolor='black', linewidth=2)
    ax.text(milestone['year'], milestone['y']+0.3, milestone['event'], 
            ha='center', fontsize=10, fontweight='bold', rotation=45)

# Ê®ôÈ°å
ax.text(1987, 9.5, 'NLP ÊäÄË°ìÁôºÂ±ïÊ≠∑Á®ã (1950-2024)', 
        ha='center', fontsize=18, fontweight='bold')

# X Ëª∏Ê®ôÁ±§
ax.text(1987, 0.3, 'Âπ¥‰ªΩ', ha='center', fontsize=12)
for year in range(1950, 2030, 10):
    ax.text(year, 0.1, str(year), ha='center', fontsize=10)

plt.tight_layout()
plt.show()

print("\nüìà NLP ÊäÄË°ìÊºîÈÄ≤Ë∂®Âã¢:")
print("  1. ÂæûË¶èÂâáÈ©ÖÂãïÂà∞Êï∏ÊìöÈ©ÖÂãï")
print("  2. Âæû‰∫∫Â∑•ÁâπÂæµÂà∞Ëá™ÂãïÂ≠∏ÁøíÁâπÂæµ")
print("  3. ÂæûÂ∞èÊ®°ÂûãÂà∞Â§ßÂûãË™ûË®ÄÊ®°Âûã")
print("  4. ÂæûÂñÆ‰∏Ä‰ªªÂãôÂà∞ÈÄöÁî®Êô∫ËÉΩ")
````

````python
# Áπ™Ë£ΩÊäÄË°ìÊºîÈÄ≤ËàáÈóúËÅØÂúñ
import networkx as nx

# ÂâµÂª∫ÊúâÂêëÂúñ
G = nx.DiGraph()

# Ê∑ªÂä†ÁØÄÈªû (ÊäÄË°ì)
technologies = [
    ('Ë¶èÂâáÁ≥ªÁµ±', 1990, 'coral'),
    ('Áµ±Ë®àML', 2000, 'skyblue'),
    ('Word2Vec', 2013, 'lightgreen'),
    ('RNN', 2014, 'gold'),
    ('LSTM', 2015, 'orange'),
    ('Seq2Seq', 2016, 'plum'),
    ('Attention', 2016, 'pink'),
    ('Transformer', 2017, 'red'),
    ('BERT', 2018, 'lightcoral'),
    ('GPT-2', 2019, 'mediumpurple'),
    ('GPT-3', 2020, 'mediumorchid'),
    ('ChatGPT', 2022, 'violet'),
]

for tech, year, color in technologies:
    G.add_node(tech, year=year, color=color)

# Ê∑ªÂä†ÈÇä (ÊºîÈÄ≤Èóú‰øÇ)
edges = [
    ('Ë¶èÂâáÁ≥ªÁµ±', 'Áµ±Ë®àML'),
    ('Áµ±Ë®àML', 'Word2Vec'),
    ('Word2Vec', 'RNN'),
    ('RNN', 'LSTM'),
    ('LSTM', 'Seq2Seq'),
    ('Seq2Seq', 'Attention'),
    ('Attention', 'Transformer'),
    ('Word2Vec', 'Transformer'),  # Ë©ûÂêëÈáè‰πüÂΩ±Èüø Transformer
    ('Transformer', 'BERT'),
    ('Transformer', 'GPT-2'),
    ('GPT-2', 'GPT-3'),
    ('GPT-3', 'ChatGPT'),
    ('BERT', 'ChatGPT'),  # BERT ÁöÑÈ†êË®ìÁ∑¥ÊÄùÊÉ≥‰πüÂΩ±Èüø GPT
]

G.add_edges_from(edges)

# Â∏ÉÂ±Ä
pos = nx.spring_layout(G, k=2, iterations=50, seed=42)

# Áπ™Âúñ
plt.figure(figsize=(16, 12))
node_colors = [G.nodes[node]['color'] for node in G.nodes()]

nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=3000, alpha=0.9, edgecolors='black', linewidths=2)
nx.draw_networkx_labels(G, pos, font_size=11, font_weight='bold')
nx.draw_networkx_edges(G, pos, edge_color='gray', arrows=True, arrowsize=20, arrowstyle='->', width=2, alpha=0.6)

# Ê∑ªÂä†Âπ¥‰ªΩÊ®ôË®ª
for node, (x, y) in pos.items():
    year = G.nodes[node]['year']
    plt.text(x, y-0.08, f'({year})', ha='center', fontsize=9, style='italic', color='darkblue')

plt.title('NLP ÊäÄË°ìÊºîÈÄ≤ËàáÈóúËÅØÂúñ (1990-2024)', fontsize=18, fontweight='bold', pad=20)
plt.axis('off')
plt.tight_layout()
plt.show()

print("\nüîó ÊäÄË°ìÊºîÈÄ≤ÈóúÈçµÁØÄÈªû:")
print("  1. Word2Vec (2013): ÈñãÂïüË©ûÂêëÈáèÊôÇ‰ª£")
print("  2. Attention (2016): ÁÇ∫ Transformer Â•†ÂÆöÂü∫Á§é")
print("  3. Transformer (2017): ÂæπÂ∫ïÊîπËÆä NLP ÁØÑÂºè")
print("  4. BERT (2018): È†êË®ìÁ∑¥Ê®°ÂûãÁöÑÈáåÁ®ãÁ¢ë")
print("  5. GPT-3 (2020): Â±ïÁ§∫Â§ßÊ®°ÂûãÁöÑÂº∑Â§ßËÉΩÂäõ")
print("  6. ChatGPT (2022): AI Ëµ∞ÂêëÂ§ßÁúæÊáâÁî®")
````
