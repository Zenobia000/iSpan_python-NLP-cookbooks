# 牛刀小試

```python
# 用來進行 tokenize, part_of_speech 和 named entity recognition 的 文本 (這邊使用英文歌詞)
lyrics = '''
When I was young,
I'd listen to the radio,
Waitin' for my favorite songs,
When they played I'd sing along,
It made me smile.
'''
```

```python
lyrics_v2 = '''
Apple Inc. is planning to open a new store in Paris, France. The company's CEO, Tim Cook, announced the expansion during a recent conference in San Francisco. The store will be located on Champs-Élysées, one of the most famous avenues in the world.

Apple's new store will feature the latest products, including the iPhone 13 and the MacBook Pro with M2 chip. It will also offer technical support and workshops for customers.

Paris is known for its rich history and culture. The Louvre Museum, Eiffel Tower, and Notre-Dame Cathedral are some of the city's iconic landmarks. Tourists from all over the globe visit Paris to experience its art, cuisine, and fashion.

The Champs-Élysées store is part of Apple's global expansion strategy, which aims to bring its innovative technology to more people around the world. Apple Inc. has a strong presence in major cities across Europe, Asia, and North America.

Tim Cook expressed excitement about the new venture and said, "We are thrilled to bring the magic of Apple to the heart of Paris. Our goal is to enhance the lives of our customers through our products and services."

The opening date of the new Apple store in Paris is set for next month, and anticipation is building among tech enthusiasts and Parisians alike.

'''
```

```python
# 匯入套件
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
from nltk import ne_chunk

# ne_chunk 會用到 numpy
import numpy as np
import pandas as pd
```

```python
# 下載文句分詞器 (sentence tokenizer)
nltk.download('punkt')

# 下載詞性標註器 (tagger)
nltk.download('averaged_perceptron_tagger')

# 下載命名實體辨識工具
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
# 分詞
tokens = word_tokenize(lyrics_v2); tokens
```

```python
# 詞性標註 (part-of-speech)
pos_tags = pos_tag(tokens); pos_tags
```

```python
# 命名實體辨識
ne_chunks = ne_chunk(pos_tags); ne_chunks
```

```python
# NER NLTK 提高辨識能力: https://segmentfault.com/a/1190000017015436
# extract all named entities
named_entities = []

for tagged_tree in ne_chunks:
    if hasattr(tagged_tree, "label"):
        entity_name = tagged_tree.leaves()[0][0]
        entity_type = tagged_tree.label()
        named_entities.append((entity_name, entity_type))

entity_frame = pd.DataFrame(named_entities, columns=['Entity Name', 'Entity Type']); entity_frame
```
