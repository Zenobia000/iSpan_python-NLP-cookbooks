# 使用 re 做英文分詞

'\s': 空白字元（空白、換行、tab等）

# 抓取特定資訊
回傳所有符合正規表示式的字串：re.findall(pattern, string) <br>

## 比對漢字（含繁簡中文、日本漢字、韓國漢字等）
中日韓統一表意文字 Unicode 範圍：4E00-9FA5 <br>
(Reference: https://zh.wikipedia.org/wiki/中日韓統一表意文字)

```python
re.findall('[\u4E00-\u9FA5]+', text_tw)
```

## 比對英文單字
含大小寫英文及連字號（hyphen）'-' <br>
e.g. heart-rending

```python
re.findall('[A-Za-z-]+', text_en)
```

# 比對特定詞
### 比對中文引號內句子

```python
re.findall('「[\u4E00-\u9FA5，。]+」', text_tw)
```

### 比對英文引號內句子

```python
re.findall('“[A-Za-z0-9’,. ]+”',  text_en)
```

### 比對大學名稱

```python
re.findall('[A-Za-z]+ University', text_en)
```

### 比對特定日期格式：月、日、年

```python
re.findall('[A-Z][a-z]+. [0-9]+, [0-9]+',  text_en)
```

```python
re.findall('([A-Z][a-z]+). ([0-9]+), ([0-9]+)',  text_en)
```

```python
re.findall('[0-9]+年[0-9]+月[0-9]+日',  text_tw)
```

### 比對領域學家

```python
re.findall('[A-Za-z]+ist[ s]', text_en)
```

### 比對先生/女士名

```python
re.findall('Ms. [A-Z][a-z]+|Mr. [A-Z][a-z]+', text_en)
```

## 回傳比對字串位置
re.finditer(pattern, string): return re.Match object <br>
- Match.start() 
- Match.end()

```python
[(match.start(), match.end()) for match in re.finditer('[A-Za-z]+ University', text_en)]
```

## 使用()比對子字串
- 完整字串：Match.group()
- 子字串n：Match.group(n)

```python
[(match.group()) for match in re.finditer('([A-Z][a-z]+). ([0-9]+), ([0-9]+)', text_en)]
```

```python
[(match.group(1), match.group(2), match.group(3)) for match in re.finditer('([A-Z][a-z]+). ([0-9]+), ([0-9]+)', text_en)]
```
