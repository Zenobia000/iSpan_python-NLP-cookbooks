# 讀取新聞內容

```python
with open ('ny_news_zh_tw.txt', 'r') as f:
    text = f.read()
print(text)
```

# 使用 jieba，並切換至繁體中文詞典
使用自行定義詞典：dict.txt.big.tw

```python
import jieba
jieba.load_userdict("dict.txt.big.tw")
```

# 分詞（Tokenization）
精準模式（預設）：jieba.cut(text)

```python
words = jieba.cut(text)
print("/".join(words))
```

# 移除標點符號

```python
punctuation = set()
with open ('punctuation_zh_tw.txt', 'r') as f:
    for line in f.readlines():
        punctuation.add(line.strip())
with open ('punctuation_en.txt', 'r') as f:
    for line in f.readlines():
        punctuation.add(line.strip())
punctuation.add('\n')
punctuation.add(' ')
print(punctuation)
```

```python
words_list = [word for word in jieba.cut(text) if word not in punctuation]
```

```python
words_list
```

# WordCount

```python
word_count = dict()
for word in words_list:
    if(word not in word_count.keys()):
        word_count.update({word:1})
    else:
        word_count.update({word:word_count[word]+1})
```

```python
word_count
```

```python
words_dic = {k: v for k, v in sorted(word_count.items(), key=lambda item: item[1], reverse=True)}
words_dic
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family']='simhei'
plt.figure(figsize=(15,6))
plt.xticks(rotation='vertical', size=14)
plt.yticks(size=14)
plt.bar(list(words_dic.keys())[:50], list(words_dic.values())[:50] ,align='center')
```

# 移除停止詞（stop words）

```python
stopwords = set()
with open ('stopwords_zh_tw.txt', 'r') as f:
    for line in f.readlines():
        stopwords.add(line.strip())
print(stopwords)
```

```python
words_list = [word for word in jieba.cut(text) if word not in punctuation and word not in stopwords]
```

```python
word_count = dict()
for word in words_list:
    if(word not in word_count.keys()):
        word_count.update({word:1})
    else:
        word_count.update({word:word_count[word]+1})
```

```python
words_dic = {k: v for k, v in sorted(word_count.items(), key=lambda item: item[1], reverse=True)}
words_dic
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family']='simhei'
plt.figure(figsize=(15,6))
plt.xticks(rotation='vertical', size=14)
plt.yticks(size=14)
plt.bar(list(words_dic.keys())[:50], list(words_dic.values())[:50] ,align='center')
```

## 加入自定義詞的三種方法：
1. jieba.add_word(new_word, freq=freq) <br>
新詞的 freq 預設在 add_word 後是 1，可自行設定，freq 越高被切出來的機率越高。
2. jieba.suggest_freq(word)<br>
suggest_freq 可使該字 freq+1
3. 手動新增至詞典中(dict.txt.big.tw)並定義詞性<br>
新南威爾斯州 1 n

### 範例1：jieba.add_word(new_word)

```python
string = '新南威爾斯州可能失去了四分之一的無尾熊。大量的澳棕短鼻袋狸和長足的長鼻袋鼠很有可能已經死亡，後者是一種沙袋鼠'
```

```python
words = jieba.cut(string)
print(list(words))
```

```python
jieba.add_word('新南威爾斯州')
jieba.add_word('澳棕短鼻袋狸')
jieba.add_word('長鼻袋鼠')
jieba.add_word('沙袋鼠')
#刪除詞：jieba.del_word('word')
```

```python
words = jieba.cut(string)
print(list(words))
```

### 範例2：jieba.add_word(word, freq)

```python
words = jieba.cut('全台大停電')
print(list(words))
```

```python
jieba.add_word('全台', freq=10)
```

```python
words = jieba.cut('全台大停電')
print(list(words))
```

```python
jieba.add_word('全台', freq=40)
```

### 範例3：jieba.suggest_freq(word, tune=True)

```python
words = jieba.cut('台中天氣很好')
print(list(words))
```

```python
jieba.suggest_freq('台中', tune=True)
```

```python
words = jieba.cut('台中天氣很好')
print(list(words))
```

# 以 HMM（隱馬可夫模型）演算法發現新詞

```python
words = jieba.cut('卜松是一間好企業')
print(list(words))
```

```python
words = jieba.cut('卜松是一間好企業', HMM=False)
print(list(words))
```
