# 準備資料

# 讀取新聞內容

```python
with open ('ny_news_en.txt', 'r') as f:
    text = f.read()
print(text)
```

# 分詞（Tokenization）
根據空白字元來分詞：str.spilt()

```python
words = text.split()
print("/ ".join(words))
```

```python
import nltk
nltk.download('punkt')
```

```python
words = nltk.word_tokenize(text)
print("/".join(words))
```

# 移除標點符號

```python
punctuation = set()
with open ('punctuation_en.txt', 'r') as f:
    for line in f.readlines():
        punctuation.add(line.strip())
print(punctuation)
```

```python
words_list = [word for word in words if word not in punctuation]
```

```python
words_list
```

# 詞形還原（Lemmatization）
需下載語料庫：nltk.download('averaged_perceptron_tagger') <br>
詞性判斷（依上下文判斷）：nltk.pos_tag(words_list, pos=pos) <br>
pos 設定為：n(名詞), v(動詞), a(形容詞) <br>
##### 注意：開頭若是大寫將無法正確轉換，如：lemmatizer.lemmatize('Animals', pos='n') => Animals

```python
nltk.download('averaged_perceptron_tagger')
```

```python
words_tag = nltk.pos_tag(words_list)
words_tag
```

## 句子開頭大寫轉小寫：
判斷標準：詞性屬於專有名詞 NNP/NNPS（proper noun）以外，皆轉為小寫：str.lower() <br>
備註：少數人名無法被 NLTK 判斷為 NNP/NNPS、少數開頭大寫的名詞被誤判為專有名詞

```python
for i in range(len(words_tag)):
    if(words_tag[i][1]!='NNP' and words_tag[i][1]!='NNPS'):
        words_list[i] = words_list[i].lower()
    else:
        continue
```

```python
words_list
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer() 
nltk.download('wordnet')
```

```python
for i in range(len(words_tag)):
    if(words_tag[i][1][0]=='N'): #名詞
        words_list[i] = lemmatizer.lemmatize(words_list[i], pos='n')
    elif(words_tag[i][1][0]=='V'): #動詞
        words_list[i] = lemmatizer.lemmatize(words_list[i], pos='v')
    elif(words_tag[i][1][0]=='J'): #形容詞
        words_list[i] = lemmatizer.lemmatize(words_list[i], pos='a')
    else:
        continue
```

```python
words_list
```

# WordCount

```python
word_count = dict()
for word in words_list:
    if(word not in word_count.keys()):
        word_count.update({word:1})
    else:
        word_count.update({word:word_count[word]+1})
```

```python
word_count
```

```python
words_dic = {k: v for k, v in sorted(word_count.items(), key=lambda item: item[1], reverse=True)}
words_dic
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,6))
plt.xticks(rotation='vertical', size=14)
plt.yticks(size=14)
plt.bar(list(words_dic.keys())[:50], list(words_dic.values())[:50], align='center')
```

# 移除停止詞（stop words）
1. 使用 nltk 內建 stopwords
2. 使用自定義 stopwords（stopwords_en.txt）

```python
nltk.download('stopwords')

from nltk.corpus import stopwords
set(stopwords.words('english'))
```

```python
stopwords = set()
with open ('stopwords_en.txt', 'r') as f:
    for line in f.readlines():
        stopwords.add(line.strip())
print(stopwords)
```

```python
words_list = [word for word in words_list if word not in punctuation and word not in stopwords]
```

```python
word_count = dict()
for word in words_list:
    if(word not in word_count.keys()):
        word_count.update({word:1})
    else:
        word_count.update({word:word_count[word]+1})
```

```python
words_dic = {k: v for k, v in sorted(word_count.items(), key=lambda item: item[1], reverse=True)}
words_dic
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,6))
plt.xticks(rotation='vertical', size=14)
plt.yticks(size=14)
plt.bar(list(words_dic.keys())[:50], list(words_dic.values())[:50], align='center')
```
