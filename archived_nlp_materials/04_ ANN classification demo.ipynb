```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
```

```python
# 生成随机数据
N = 100
X = np.random.random(N) * 6 - 3
y = (X > 0).astype(int)  # 创建一个二元分类任务，如果 X 大于 0 则标签为 1，否则为 0

# 创建神经网络模型
model = keras.Sequential([
    layers.Input(shape=(1,)),  # 输入层
    layers.Dense(32, activation='sigmoid'),  # 隱藏層1 : sigmoid
    layers.Dense(16, activation='sigmoid'),  # 隱藏層2 : sigmoid
    layers.Dense(1, activation='sigmoid')  # 输出层，使用 sigmoid 激活函数进行二元分类
])

# 编译模型
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# 将数据分成训练集和测试集
X_train, y_train = X[:80], y[:80]
X_test, y_test = X[80:], y[80:]

# 训练模型，并同时记录训练集和测试集的损失和准确度
history = model.fit(X_train, y_train, epochs=200, verbose=1, validation_data=(X_test, y_test))

# 使用模型进行预测
X_plot = np.linspace(-3, 3, 100)
y_pred = model.predict(X_plot)

# 打印模型参数
model.summary()


```

```python
# 绘制训练和测试损失
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Test Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# 绘制模型的预测结果
plt.scatter(X_train, y_train, label='train Data')
plt.scatter(X_test, y_test, label='val Data')
plt.plot(X_plot, y_pred, label='Predictions', color='r')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```
