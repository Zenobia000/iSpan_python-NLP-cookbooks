## 製作詞袋（bag of words, BOW）

```python
punctuation = set()
with open ('punctuation_zh_tw.txt', 'r') as f:
    for line in f.readlines():
        punctuation.add(line.strip())
with open ('punctuation_en.txt', 'r') as f:
    for line in f.readlines():
        punctuation.add(line.strip())
punctuation.add('\n')
punctuation.add(' ')
punctuation.add('　') #全形空白
punctuation.add(u'\u00A0') #不換行空格（Non-breaking space）

stopwords = set()
with open ('stopwords_zh_tw.txt', 'r') as f:
    for line in f.readlines():
        stopwords.add(line.strip())
```

```python
bow = set()
doc_bow = dict()
for i in range(len(doc_list)):
    words = [word for word in jieba.cut(doc_list[i]) if word not in punctuation and word not in stopwords]
    doc_bow.update({i:set(words)})
    bow = bow.union(set(words))
```

# 計算自己的 IDF

```python
def compute_idf(doc_list, bow):
    import math
    import jieba
    jieba.load_userdict("dict.txt.big.tw")    
    
    # 計算每個 word 的 idf
    N = len(doc_list)
    with open('jieba/analyse/my_idf.txt','w') as f:
        for term in bow:
            dfx = 0
            for i in range(len(doc_list)):
                if term in doc_bow[i]:
                    dfx += 1
            
            idf = math.log(N/dfx ,10)
            f.write(term + ' ' + str(idf) + '\n')
```

```python
compute_idf(doc_list, bow)
```
