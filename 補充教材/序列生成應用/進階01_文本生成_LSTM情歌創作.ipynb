# 資料準備
歌手：梁靜茹、光良、楊丞琳、王心凌共 992 首歌詞

抓取資料夾中所有歌詞檔名

```python
import os
files = [file for file in os.listdir("情歌歌詞") if file.endswith(".txt")]
print(len(files))
```

讀取成一個 string

```python
text = ''
for file in files:
    print(file)
    with open('情歌歌詞/' + file, 'r') as f:
        text += f.read() + '\n'
```

```python
print('字元數:', len(text))
print(text)
```

# 開始訓練
修改自 keras 官方範例程式：https://keras.io/examples/generative/lstm_character_level_text_generation/

### 建立"字對 index"、"index 對字"的對應字典

```python
chars = sorted(list(set(text)))
char_indices = dict((c, i) for i, c in enumerate(chars))
indices_char = dict((i, c) for i, c in enumerate(chars))
```

```python
print(char_indices)
```

### 準備訓練資料
maxlen: 每筆訓練文字序列的長度 <br>
step: 這一筆跟下一筆訓練文字資料的位移 <br>
sentences: 產生的文字序列 <br>
next_chars: 文字序列的下一個"字"

```python
maxlen = 30
step = 3
sentences = []
next_chars = []
for i in range(0, len(text) - maxlen, step):
    sentences.append(text[i: i + maxlen])
    next_chars.append(text[i + maxlen])
print('產生文字序列筆數:', len(sentences))
```

```python
print(sentences[0])
```

```python
print(next_chars[0])
```

### 向量化（one-hot encoding）
輸入一個文字序列(x)，預測下一個字是什麼(y) <br>
x: (文字數列數/訓練資料數, 文字序列長度, 總字元數) <br>
y: (文字數列數/訓練資料數, 總字元數)

```python
from tensorflow.keras.callbacks import LambdaCallback
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM, GRU
from tensorflow.keras.optimizers import RMSprop
import numpy as np
import random
import sys
```

```python
x = np.zeros((len(sentences), maxlen, len(chars)), dtype=np.bool)
y = np.zeros((len(sentences), len(chars)), dtype=np.bool)
for i, sentence in enumerate(sentences):
    for t, char in enumerate(sentence):
        x[i, t, char_indices[char]] = 1
    y[i, char_indices[next_chars[i]]] = 1
```

# 建立 RNN 模型（可使用 RNN/LSTM/GRU）

```python
model = Sequential()
#model.add(LSTM(128, input_shape=(maxlen, len(chars))))
model.add(GRU(128, input_shape=(maxlen, len(chars))))
model.add(Dense(len(chars), activation='softmax'))
optimizer = RMSprop(learning_rate=0.01)
model.compile(loss='categorical_crossentropy', optimizer=optimizer)
```

```python
model.summary()
```

# 採樣方法
temperature: 文字產生的隨機性，越大越隨機、越小越固定

```python
def sample(preds, temperature=1.0):
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)
```

# 每個 epoch 訓練結束後產生文字
num_generated_chars: 產生文字數 <br>
temperautre_list: 使用的 temperature

```python
def on_epoch_end(epoch, _):
    print()
    print('#Epoch: %d' % epoch)

    num_generated_chars = 60
    temperautre_list = [0.2, 0.5, 1.0, 1.2]
    
    start_index = random.randint(0, len(text) - maxlen - 1)
    for temperature in temperautre_list:
        print('\n#temperature:', temperature)

        generated = ''
        sentence = text[start_index: start_index + maxlen]
        generated += sentence
        print('----- 起始文字 -----')
        sys.stdout.write(generated)
        
        print('\n----- 接續文字生成 -----')
        for i in range(num_generated_chars):
            x_pred = np.zeros((1, maxlen, len(chars)))
            for t, char in enumerate(sentence):
                x_pred[0, t, char_indices[char]] = 1.

            preds = model.predict(x_pred, verbose=0)[0]
            next_index = sample(preds, temperature)
            next_char = indices_char[next_index]

            sentence = sentence[1:] + next_char

            sys.stdout.write(next_char)
            sys.stdout.flush()
        print()
```

```python
print_callback = LambdaCallback(on_epoch_end=on_epoch_end)
```

# 開始訓練

```python
model.fit(x, y, batch_size=128, epochs=10, callbacks=[print_callback])
```

# 使用訓練好的模型產生文字
