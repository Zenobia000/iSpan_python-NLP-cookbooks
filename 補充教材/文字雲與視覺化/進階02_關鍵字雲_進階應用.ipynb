# 匯入 TED 資料
資料來源：https://wit3.fbk.eu/mono.php?release=XML_releases&tinfo=cleanedhtml_ted <br>
收錄各國語言的 TED 演講

```python
import xml.etree.ElementTree as et
tree = et.ElementTree(file='ted_zh-tw-20160408.xml')
```

```python
doc_list = []

for elem in tree.iter(tag='content'):
    doc_list.append(elem.text)
```

# 製作詞頻文字雲
WordCloud(font_path, background_color, scale, max_font_size, max_words).generate(string)<br>
- <b>string: 輸入以空白為間隔的word string</b><br>
- font_path: 字體路徑 <br>
- background_color: 背景顏色 <br>
- scale: 解析度 <br>
- max_font_size: 最大字體大小<br>
- max_words: 最多顯示幾個字<br>

### 計算詞頻

```python
import jieba
jieba.load_userdict("dict.txt.big.tw")
```

```python
punctuation = set()
with open ('punctuation_zh_tw.txt', 'r') as f:
    for line in f.readlines():
        punctuation.add(line.strip())
with open ('punctuation_en.txt', 'r') as f:
    for line in f.readlines():
        punctuation.add(line.strip())
punctuation.add('\n')
punctuation.add(' ')

stopwords = set()
with open ('stopwords_zh_tw.txt', 'r') as f:
    for line in f.readlines():
        stopwords.add(line.strip())
```

```python
words = [word for word in jieba.cut(doc_list[1923]) if word not in punctuation and word not in stopwords]
```

```python
words_str = " ".join(words)
words_str
```

### 產生文字雲
pip install wordcloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
wordcloud = WordCloud(font_path='jf-openhuninn-1.0.ttf', max_font_size=100, max_words=30, background_color="white",
                     scale=10).generate(words_str)
plt.figure(figsize=(10,6))
plt.imshow(wordcloud)
plt.axis("off") #不要顯示x,y軸
plt.show()
```

# 製作關鍵字雲
WordCloud().generate_from_frequencies(dictionary of keywords) <br>
- <b>dictionary of keywords: {'keyword1': 'weight1', 'keyword2': 'weight2',...}</b><br> 
- font_path: 字體路徑 <br>
- background_color: 背景顏色 <br>
- scale: 解析度 <br>
- max_font_size: 最大字體大小<br>
- max_words: 最多顯示幾個字<br>

### 計算 TF-IDF

```python
import jieba
import jieba.analyse
# 使用自己的 IDF 文件路徑
jieba.analyse.set_idf_path("jieba/analyse/my_idf.txt")
```

```python
keywords = jieba.analyse.extract_tags(doc_list[1923], topK=20, withWeight=True, allowPOS=('n','nr','ns','nz','nt'))
```

```python
keywords
```

```python
#轉成 dictionary
dict(keywords)
```

### 產生關鍵字雲

```python
wordcloud = WordCloud(font_path='jf-openhuninn-1.0.ttf', max_font_size=100, max_words=20, background_color="white",
                     scale=10).generate_from_frequencies(dict(keywords))
plt.figure(figsize=(10,6))
plt.imshow(wordcloud)
plt.axis("off") #不要顯示x,y軸
plt.show()
```
