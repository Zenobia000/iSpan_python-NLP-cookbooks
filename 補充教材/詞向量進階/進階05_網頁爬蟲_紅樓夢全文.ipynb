# 一次爬取紅樓夢全文共120回
time.sleep(n): 休息 n 秒後再繼續爬取，避免 access denied

```python
import time
for n in range(1,121):
    with open('紅樓夢全文/ch'+ str(n) + '.txt', 'w') as f:
        # 爬取 Chapter n 文字
        print('爬取紅樓夢第'+ str(n) +'回')
        res = requests.get('https://ctext.org/hongloumeng/ch'+ str(n) +'/zh')
        soup = BeautifulSoup(res.text,'html.parser')
    
        for i in range(len(soup.select("td[class='ctext']"))):
            f.write(soup.select("td[class='ctext']")[i].text.strip() + '\n')
    
    time.sleep(3)
```
