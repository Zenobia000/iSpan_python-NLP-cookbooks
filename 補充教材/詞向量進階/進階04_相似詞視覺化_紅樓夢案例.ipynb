# 讀取紅樓夢全文
共120回

```python
corpus = []
for i in range(1,121):
    with open('紅樓夢全文/ch'+ str(i) + '.txt', 'r') as f:
        corpus.append(f.read())
```

# 中文預處理
只取名詞、動詞、形容詞

```python
import jieba
jieba.load_userdict("dict.txt.big.tw")
```

```python
punctuation = set()
stopwords = set()
with open ('punctuation_zh_tw.txt', 'r') as f:
    for line in f.readlines():
        punctuation.add(line.strip())
with open ('punctuation_en.txt', 'r') as f:
    for line in f.readlines():
        punctuation.add(line.strip())
punctuation.add('\n')
punctuation.add(' ')

with open ('stopwords_zh_tw.txt', 'r') as f:
    for line in f.readlines():
        stopwords.add(line.strip())
```

# 匯入紅樓夢人物列表
- 加入詞典：jieba.add_word(word, freq, tag)

```python
characters = []
with open ('紅樓夢人物列表.txt', 'r') as f:
    characters = f.readline().split(' ')
```

```python
characters
```

```python
for word in characters:
    jieba.add_word(word, freq=2, tag='nr')
```

# 中文分詞
可再自行定義詞典（文言文），分詞效果應會更佳

```python
import jieba.posseg as pseg
doc_list = []
for doc in corpus:
    words_list = []
    for word, tag in pseg.cut(doc):
        if(tag[0]=='n' or tag[0]=='v' or tag[0]=='a'):
            words_list.append(word)
    
    words_list = [word for word in words_list if word not in punctuation and word not in stopwords]
    doc_list.append(words_list)
```

# word2vec 建模
class gensim.models.word2vec.Word2Vec(sentences=None, size=100, window=5, min_count=5, max_vocab_size=None, sample=0.001, seed=1, workers=3, min_alpha=0.0001, sg=0, hs=0, negative=5, ns_exponent=0.75, cbow_mean=1, hashfxn=<built-in function hash>, iter=5, null_word=0, trim_rule=None, sorted_vocab=1, batch_words=10000, compute_loss=False, callbacks=(), max_final_vocab=None)
    
- sentences: 分詞 list 的 document list
- size: 向量維度
- window: 視窗大小
- min_count: 最小詞頻，忽視低於此詞頻的詞
- seed: 隨機種子
- workers: 執行緒（thread）的數量，對於多核心的CPU有加速功能
- sg: 1 for skip-gram; 0 for CBOW.
- hs: 1 for hierarchical softmax; 0 for negative sampling.
- negative: 負採樣數量（通常設定 5-20）
- ns_exponent: 採樣機率調整參數，建議值 0.75
- max_vocab_size: 最大單詞量

document: https://radimrehurek.com/gensim/models/word2vec.html

```python
from gensim.models import word2vec
model = word2vec.Word2Vec(doc_list, size=100, window=5, seed=0, workers=4, sg=1, hs=0)
```

### 根據 Embedding Projector 格式要求輸出 vectors 和 words

```python
with open('word2vec_vectors.tsv', 'w') as f:
    for word in model.wv.vocab.keys():
        vector_str = ''
        for element in model.wv[word]:
            vector_str += str(element) + '\t'
        vector_str = vector_str.strip()
        f.write(vector_str + '\n')
```

```python
with open('word2vec_words.tsv', 'w') as f:
    for word in model.wv.vocab.keys():
        f.write(word + '\n')
```
